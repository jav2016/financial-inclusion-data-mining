# Week 1 Lab: Diabetes Dataset Exploration

This notebook follows the Week 1 Python idea: use simple code to load data, print information, and understand the dataset.

Dataset file used in this notebook: `diabetes_binary_health.csv`

## Step 1: Import Libraries

We use `pandas` to read and explore the CSV file. We use `matplotlib` to draw simple charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2: Load the CSV File

The CSV file is saved in the same folder as this notebook, so we only need the file name.

```python
df = pd.read_csv("diabetes_binary_health.csv")

df.head()
```

## Step 3: Check Dataset Size

`shape` shows how many rows and columns are in the dataset.

```python
df.shape
```

## Step 4: View Column Names

```python
df.columns
```

## Step 5: Dataset Information

`info()` shows column names, data types, and whether values are missing.

```python
df.info()
```

## Step 6: Check Missing Values

```python
df.isnull().sum()
```

## Step 7: Basic Statistics

`describe()` gives count, mean, minimum, maximum, and other summary values.

```python
df.describe()
```

## Step 8: Count Diabetes and No Diabetes Records

`Diabetes_binary` is the main column for our final project.

- `0` = no diabetes
- `1` = diabetes

```python
df["Diabetes_binary"].value_counts()
```

## Step 9: Diabetes Count Chart

```python
df["Diabetes_binary"].value_counts().plot(kind="bar")

plt.title("Diabetes and No Diabetes Count")
plt.xlabel("Diabetes_binary")
plt.ylabel("Number of People")
plt.show()
```

## Step 10: Average Values by Diabetes Status

This helps us compare people with diabetes and people without diabetes.

```python
df.groupby("Diabetes_binary").mean()
```

## Step 11: Compare BMI by Diabetes Status

```python
df.groupby("Diabetes_binary")["BMI"].mean()
```

## Step 12: BMI Chart

```python
df.groupby("Diabetes_binary")["BMI"].mean().plot(kind="bar")

plt.title("Average BMI by Diabetes Status")
plt.xlabel("Diabetes_binary")
plt.ylabel("Average BMI")
plt.show()
```
